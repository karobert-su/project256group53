```python
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob
```

```python
from textblob import TextBlob
```

```python
url = input("Please enter your URL to your text: ")
```

```python
#gathers whole textbook
response = requests.get(url)
text = response.text
print(text)
```

```python
soup = BeautifulSoup(text)

print(soup.text)
```

```python
blob = TextBlob(soup.text)

blob.noun_phrases
```
